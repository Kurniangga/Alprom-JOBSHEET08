Praktikum 07 – Fungsi Decorator

```python
def decorator_sapa (fungsi) :
    def wrapper() :
        print("Sebelum fungsi dijalankan...")
        fungsi()
        print ("Setelah fungsi dijalankan...")
    return wrapper

@decorator_sapa
def sapa() :
  print ("Halo, Mahasiswa Polines !")
```

Penugasan Nomor 10 - Fungsi Decorator

```python
def decorator_halo (fungsi):
  def wrapper():
    print("Memulai fungsi...")
    fungsi()
    print("Fungsi selesai dijalankan.")
  return wrapper

@decorator_halo
def halo():
  print("Halo, Mahasiswa Polines !")

halo()
```

Penugasan Nomor 11A - Studi Kasus Nyata-Program Nilai

```python
nilai_akhir = lambda tugas, ujian: (tugas*0.4) + (ujian*0.6)

def input_nilai():
  list_nilai = []
  jumlah = int(input("Jumlah mata kuliah: "))

  for i in range(jumlah):
    print(f"\nMata Kuliah ke-{i+1}")
    tugas = float(input("Nilai tugas: "))
    ujian = float(input("Nilai ujian: "))

    nilai_jadi = nilai_akhir(tugas, ujian)
    list_nilai.append(nilai_jadi)
  return list_nilai

def hitung_rata(data_nilai):
  return sum(data_nilai) / len(data_nilai)

def status_kelulusan(rata):
  if rata >= 60:
    return "LULUS"
  else:
    return "TIDAK LULUS"

nilai_mahasiswa = input_nilai()

print("--------------------------------------")

if nilai_mahasiswa:
  rata_rata = hitung_rata(nilai_mahasiswa)
  status = status_kelulusan(rata_rata)

  print(f"Rata-rata Nilai-nilai akhir: {nilai_mahasiswa}")
  print(f"Rata-rata: {rata_rata:.2f}")
  print(f"Status kelulusan: {status}")
```
